```python
#Avearge number of resale transactions
import numpy as np
import matplotlib.pyplot as plt

#Title for the graph 
title = "Number of resale transactions registered by flat type"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()

#Read the csv file with the genfromtxt() function and extract the rows containing keywords
data = np.genfromtxt ('resale-transaction-by-flat-type-based-on-registered-cases/resale-transactions-by-flat-type-based-on-registered-cases.csv',
                     skip_header=1,dtype=[('financial_year','U50'),('flat_type', 'U50'),('resale_transactions','i4')], delimiter=",", 
                     missing_values =['na','-'],filling_values=[0])
clean_data= data[np.isin(data['financial_year'],['2015','2016','2017','2018','2019','2020'])]

# Print out total rows and columns of data in the file
print("There are {} rows and {} columns in this dataset.".format(len(clean_data),len(clean_data[0])))
print()

# Print out the unique number of financial years ,flat type and resale transactions of data captured
set_financial_year = set(clean_data['financial_year'])
set_flat_type = set(clean_data['flat_type'])
set_resale_transactions = set(clean_data['resale_transactions'])
print(str(len(set_financial_year)) + " unique values in financial_year column")
print(str(len(set_flat_type)) + " unique values in flat type column")
print(str(len(set_resale_transactions)) + " unique values in no. of resale transactions column")
print()

# Calculate and print out the year which has the highest and lowest total number of transactions
max_row_index = np.argmax(clean_data['resale_transactions'])
min_row_index = np.argmin(clean_data['resale_transactions'])
print("The highest number of resale transactions was in the {} for {} with {:.0f} of resale transactions."
      .format (clean_data[max_row_index][0],clean_data[max_row_index][1], clean_data[max_row_index][2]))
print("The lowest number of resale transactions was in the {} for {} with {:.0f} of resale transactions."
      .format (clean_data[min_row_index][0], clean_data[min_row_index][1], clean_data[min_row_index][2]))

# Calculate and print out the mean and standard deviation of total number of transactions
mean = np.mean(clean_data['resale_transactions'])
std = np.std(clean_data['resale_transactions'])
print("The mean number of the resale transactions is {:.0f}".format (mean))
print("The standard deviation is {:.0f}".format (std))
print()

#Calculate and print out the total number of transactions 
total_sum = np.sum(clean_data['resale_transactions'])
print("The total number of transactions is {:.0f}.". format(total_sum))

#Calculate and print out the number of transactions for each flat type 
labels = list(set(clean_data['flat_type']))
total = np.arange(0,len(labels))
total_values = clean_data[['flat_type','resale_transactions']]
values = total_values['resale_transactions']
sum_values = {}
for i in labels:
    valuesfortotal = values[total_values['flat_type']==i]
    total_sum = np.sum(valuesfortotal)
    print("Sum for flat " + i + " is {:.0f}".format(total_sum))
    sum_values[i] = total_sum

#Note this part is to allow the barcharts to be displayed in descending order
#Plot bar chart using the calculated sum values 
from collections import OrderedDict
from operator import itemgetter  
sum_values = OrderedDict(sorted(sum_values.items(), key = itemgetter(1), reverse = True))
plt.figure(1, figsize=(25,25))
barchart = plt.bar(list(sum_values.keys()), list(sum_values.values()), color='#c27ba0')

#Print out the label of the total number of transactions for each flat type 
labels = list(sum_values.keys())
for i in range(len(barchart)):
    bar = barchart[i]
    x,y  = bar.get_xy() #set the coordinate for the labelling
    h = bar.get_height()
    plt.text(x,h,"{:.0f}".format(list(sum_values.values())[i]),fontsize=25)

#Style the graph
plt.title('Total number of resale transactions by flat type based on registered cases',fontsize=30)
plt.ylabel('No. of resale transactions',fontsize=30)
plt.yticks(fontsize=20)
plt.xticks(total, labels, fontsize=20,rotation='vertical')

plt.show()
```
